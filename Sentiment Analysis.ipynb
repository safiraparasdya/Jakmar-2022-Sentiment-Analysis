```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import re
re.compile('<title>(.*)</title>')
```

```python
#membaca data
a = pd.read_csv('1.csv')
b = pd.read_csv('2.csv')
c = pd.read_csv('3.csv')
d = pd.read_csv('4.csv')
e = pd.read_csv('5.csv')
f = pd.read_csv('6.csv')
g = pd.read_csv('7.csv')
```

```python
all_data = pd.concat([a,b,c,d,e,f,g], ignore_index= True)
all_data
```

```python
all_data_abc = pd.concat([a,b,c], ignore_index= True)
all_data_abc
```

```python
all_data['Comment']
```

```python
#removing hashtag and @username
list_comment = []
for x in all_data['Comment']:
    x = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",x).split())
    list_comment.append(x)
print(list_comment)
```

```python
list_comment
```

```python
df = pd.DataFrame(list_comment)
df
```

```python
df.rename(columns={0: 'Comment_clean'}, inplace=True)
df
```

```python
all_data2 = pd.concat([all_data, df], axis=1)
all_data2
```

```python
all_data2.to_csv('all_data2.csv')
```
